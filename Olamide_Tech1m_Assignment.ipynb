```python
import numpy as np
import pandas as pd
```

```python
train = pd.read_csv(r"C:\Users\User\Downloads\train1.csv")
```

```python
train
```

```python
# without promo 2013-01-03
no_promo_in_2013 = train.loc[(train["Promo"]==0) & (train["Date"]=="2013-01-03")]
```

```python
no_promo_in_2013
```

```python
# with promo 2013-01-07
promo_in_2013 = train.loc[(train["Promo"]==1) & (train["Date"]=="2013-01-07")]
```

```python
promo_in_2013
```

```python
# total sum of sales and customers without promo 2013-01-03
no_promo_in_2013.groupby("Date").sum().iloc[:,2:4]
```

```python
a = no_promo_in_2013.groupby("Date").sum().iloc[:,2:4]
```

```python
# total sum of sales and customers with promo 2013-01-07
promo_in_2013.groupby("Date").sum().iloc[:,2:4]
```

```python
b = promo_in_2013.groupby("Date").sum().iloc[:,2:4]
```

```python
combined = pd.concat([a,b])
```

```python
combined
```

```python
# bar chart showing the difference(in sales and customers) between 2013-01-03(NO promo) and 2013-01-07(promo)
combined.plot(kind ='bar', linewidth =2.0 )
```
